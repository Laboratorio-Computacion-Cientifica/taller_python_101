# **Libreras (o paquetes) y Frameworks**

En programacin cuando hablamos de **librera**, nos referimos a un conjunto de archivos de cdigo que nos facilitan funcionalidades. Las libreras nos premiten **reutilizar** cdigo.

En Python a una librera, la llamamos **paquete**. Cada paquete de python esta compuesto por uno o ms archivos (**mdulos**) Python.

Cuando hablamos de **Framework**, nos referimos a un concepto ms amplio, ya no solo a funcionalidades implementadas sino a todo un marco de trabajo que puede incluir una o ms libreras y herramientas.

## **Arreglos unidimensionales**

Nuestro primer arreglo con NumPy!

```python
# arreglo de una dimensin
l = [1,2,3,4]
x = np.array(l)
print(x)
print(type(x))
```

```python
x = np.arange(10)
print(x)
```

### **Arreglos n-dimensionales**

```python
# arreglo bidimensional
l = [[1,2,3],[4,5,6]]  #[[1,2,3],[4,5,6]]
print(len(l[0]))
x = np.array(l)
print(x)
print(type(x))
print(x.shape)

print(x*x)
```

```python
# arreglo multidimensional (matriz)
x = np.array([[[1,2,3]],[[4,5,6]],[[4,5,6]]])
#print(x)
#print(x.shape) # resultado(3,1,3) --> (cant_elementos, filas, columnas)

print(x*x)
```

### **Arreglos + tipos de datos**

```python
# matriz de enteros
x = np.matrix([0,2,3], dtype=int) # dtype = data type
print(x)
print(type(x))
```

```python
# matriz de decimales
x = np.matrix([0,2,3], dtype=float)
print(x)
```

```python
# matriz booleana
x = np.matrix([0,2,3], dtype=bool)
print(x)
print(1*x)
```

### **Atributos de arreglos**

```python
# shape, reshape, ndim, itemsize, 
x = np.matrix([0,2,3], dtype=int)
print(x.shape)
```

## **Redimensionar arreglos**

```python
x = np.matrix([0,2,3,4], dtype=int)
print(x)
print(x.shape)

x2 = x.reshape(2,2)
print(x2)
print(x2.shape)
```

```python
# matriz de nxm de unos
x = np.ones([2,3], dtype = int)
print(x)
print(x.shape)
```

## **Operaciones**
Aqu se puede observar claramente la diferencia con el tipo de datos "lista"

```python
# arreglo bidimensional
l = [[1,2,3],[4,5,6]]
x = np.array(l)
print(x)
```

```python
print(x.max())
```

```python
print(x.min())
```

```python
print(x.mean())
```

```python
# operaciones aritmticas: producto
c = np.array([[4, 3], [2, 1]])
d = np.array([[1, 2], [3, 4]])
print(c)
print("---")
print(d)
print(type(c))
```

```python
# producto elemento a elemento
print(c*d)
print("---")
# producto matricial
print(c@d)
```

```python
# producto con tipo de dato "matrix"
c = np.matrix([[4, 3], [2, 1]])
d = np.matrix([[1, 2], [3, 4]])
print(c)
print("---")
print(d)
print(type(c))
```

```python
print(c*d)
print("----")
print(np.dot(c,d))
```

```python
a = np.array([[1,2,3]])
b = np.array([[2,3,4]])
print(np.multiply(a,b))
print(np.vdot(a,b))
```

## **Iterar arreglos**

```python
# arreglo de una dimensin
x = np.array([1,2,3,4])
print(x)

for elemento in np.ndenumerate(x):
    print(elemento)

print("---")

x = np.array([[1,2,3,4],[5,6,7,8]])
print(x)


for indice,valor in np.ndenumerate(x):
    print(indice)
    print(valor)
```

## Creamos una lista de valores aleatorios usando NumPy

## Podemos aumentar el tamao de la figura

```python
plt.figure(figsize=(15, 10))
plt.hist(valores_aleatorios)
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 2 * np.pi, 1024)
data2d = np.sin(t)[:, np.newaxis] * np.cos(t)[np.newaxis, :]

fig, ax = plt.subplots()
im = ax.imshow(data2d)
ax.set_title('Pan on the colorbar to shift the color mapping\n'
             'Zoom on the colorbar to scale the color mapping')

fig.colorbar(im, ax=ax, label='Interactive colorbar')

plt.show()
```

## Podemos agregar mas informacin a las grficas utilizando distintas funciones de la librera matplotlib

Algunas de esas son:
- `plt.title()`: Permite agregar un ttulo a nuestra figura
- `plt.xticks()`: Permite cambiar los valores que se muestran en el eje X
- `plt.yticks()`: Permite cambiar los valores que se muestran en el eje Y
- `plt.xlabel()`: Permite agregar un texto en el eje X
- `plt.ylabel()`: Permite agregar un texto en el eje Y

```python
plt.figure(figsize=(15, 10))
plt.title("Ocurrencias de numeros aleatorios")
plt.xlabel("Numeros")
plt.ylabel("Cantida de Ocurrencias")
plt.hist(valores_aleatorios)
```
