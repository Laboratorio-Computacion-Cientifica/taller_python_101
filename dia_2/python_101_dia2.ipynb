# **Librerías (o paquetes) y Frameworks**

En programación cuando hablamos de **librería**, nos referimos a un conjunto de archivos de código que nos facilitan funcionalidades. Las librerías nos premiten **reutilizar** código.

En Python a una librería, la llamamos **paquete**. Cada paquete de python esta compuesto por uno o más archivos (**módulos**) Python.

Cuando hablamos de **Framework**, nos referimos a un concepto más amplio, ya no solo a funcionalidades implementadas sino a todo un marco de trabajo que puede incluir una o más librerías y herramientas.

# **NumPy**

Usando NumPy, se pueden realizar las siguientes operaciones:

- Operaciones matemáticas y lógicas sobre matrices.

- Operaciones relacionadas con el álgebra lineal. NumPy tiene funciones integradas para álgebra lineal y generación de números aleatorios.


El objeto más importante definido en NumPy es un tipo de matriz N-dimensional llamado ndarray. Describe la colección de artículos del mismo tipo. Se puede acceder a los elementos de la colección mediante un índice de base cero.

Cada elemento en un ndarray toma el mismo tamaño de bloque en la memoria. Cada elemento en ndarray es un objeto de tipo de datos (llamado dtype).

## **Arreglos unidimensionales**

Nuestro primer arreglo con NumPy!

```python
# arreglo de una dimensión
l = [1,2,3,4]
x = np.array(l)
print(x)
print(type(x))
```

```python
x = np.arange(10)
print(x)
```

### **Arreglos n-dimensionales**

```python
# arreglo bidimensional
l = [[1,2,3],[4,5,6]]  #[[1,2,3],[4,5,6]]
print(len(l[0]))
x = np.array(l)
print(x)
print(type(x))
print(x.shape)

print(x*x)
```

```python
# arreglo multidimensional (matriz)
x = np.array([[[1,2,3]],[[4,5,6]],[[4,5,6]]])
#print(x)
#print(x.shape) # resultado(3,1,3) --> (cant_elementos, filas, columnas)

print(x*x)
```

### **Arreglos + tipos de datos**

```python
# matriz de enteros
x = np.matrix([0,2,3], dtype=int) # dtype = data type
print(x)
print(type(x))
```

```python
# matriz de decimales
x = np.matrix([0,2,3], dtype=float)
print(x)
```

```python
# matriz booleana
x = np.matrix([0,2,3], dtype=bool)
print(x)
print(1*x)
```

### **Atributos de arreglos**

```python
# shape, reshape, ndim, itemsize, 
x = np.matrix([0,2,3], dtype=int)
print(x.shape)
```

## **Redimensionar arreglos**

```python
x = np.matrix([0,2,3,4], dtype=int)
print(x)
print(x.shape)

x2 = x.reshape(2,2)
print(x2)
print(x2.shape)
```

### **Rutinas de creación particulares**

```python
# arreglo de unos
unos = np.ones(5)
print(unos)
```

```python
# arreglo de ceros
ceros = np.zeros(5)
print(ceros)
```

```python
# matriz de nxm de unos
x = np.ones([2,3], dtype = int)
print(x)
print(x.shape)
```

## **Operaciones**
Aquí se puede observar claramente la diferencia con el tipo de datos "lista"

```python
# arreglo bidimensional
l = [[1,2,3],[4,5,6]]
x = np.array(l)
print(x)
```

```python
print(x.max())
```

```python
print(x.min())
```

```python
print(x.mean())
```

```python
# operaciones aritméticas: producto
c = np.array([[4, 3], [2, 1]])
d = np.array([[1, 2], [3, 4]])
print(c)
print("---")
print(d)
print(type(c))
```

```python
# producto elemento a elemento
print(c*d)
print("---")
# producto matricial
print(c@d)
```

```python
# producto con tipo de dato "matrix"
c = np.matrix([[4, 3], [2, 1]])
d = np.matrix([[1, 2], [3, 4]])
print(c)
print("---")
print(d)
print(type(c))
```

```python
print(c*d)
print("----")
print(np.dot(c,d))
```

```python
a = np.array([[1,2,3]])
b = np.array([[2,3,4]])
print(np.multiply(a,b))
print(np.vdot(a,b))
```

## **Iterar arreglos**

```python
# arreglo de una dimensión
x = np.array([1,2,3,4])
print(x)

for elemento in np.ndenumerate(x):
    print(elemento)

print("---")

x = np.array([[1,2,3,4],[5,6,7,8]])
print(x)


for indice,valor in np.ndenumerate(x):
    print(indice)
    print(valor)
```

```python
from numpy import genfromtxt  # as lf #np.<nombre>  # np.genfromtxt
data = genfromtxt('datos.txt', delimiter=',',skip_header=1, dtype=(str,str,int,float)) #equivale a: np.genfromtxt('datos.txt', delimiter=',',skip_header=1, dtype=float)
print(data)
print(type(data))
```

## Creamos una lista de valores aleatorios usando NumPy

```python
valores_aleatorios = np.random.randint(low=1, high=10, size=100)
valores_aleatorios
```

```python
np.unique(valores_aleatorios, return_counts=True)
```

```python
from collections import Counter
frecuencias = Counter(valores_aleatorios)
frecuencias
```

## Podemos graficar un Histograma

```python
plt.hist(valores_aleatorios)
```

## Podemos aumentar el tamaño de la figura

```python
plt.figure(figsize=(15, 10))
plt.hist(valores_aleatorios)
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 2 * np.pi, 1024)
data2d = np.sin(t)[:, np.newaxis] * np.cos(t)[np.newaxis, :]

fig, ax = plt.subplots()
im = ax.imshow(data2d)
ax.set_title('Pan on the colorbar to shift the color mapping\n'
             'Zoom on the colorbar to scale the color mapping')

fig.colorbar(im, ax=ax, label='Interactive colorbar')

plt.show()
```

## Podemos agregar mas información a las gráficas utilizando distintas funciones de la librería matplotlib

Algunas de esas son:
- `plt.title()`: Permite agregar un título a nuestra figura
- `plt.xticks()`: Permite cambiar los valores que se muestran en el eje X
- `plt.yticks()`: Permite cambiar los valores que se muestran en el eje Y
- `plt.xlabel()`: Permite agregar un texto en el eje X
- `plt.ylabel()`: Permite agregar un texto en el eje Y

```python
plt.figure(figsize=(15, 10))
plt.title("Ocurrencias de numeros aleatorios")
plt.xlabel("Numeros")
plt.ylabel("Cantida de Ocurrencias")
plt.hist(valores_aleatorios)
```

## Podemos hacer un gráfico de torta

```python
from collections import Counter
frecuencias = Counter(valores_aleatorios)
frecuencias
```

```python
frecuencias.keys()
```

```python
plt.figure(figsize=(15, 10))
plt.pie(frecuencias.values(), labels=frecuencias.keys(), autopct='%1.2f%%')
```
