C1 --> C2

C1 --> C3

C2 --> C3

C3 --> C4

C4 --> C5

**Las personas dueas del emprendimiento, desean obtener la matriz que indique si existe o no el camino y cul es el inicio y el fin del recorrido**

```python
import numpy as np
```

```python
n1 = [0,1,1,0,0]
n2 = [0,0,1,0,0]
n3 = [0,0,0,1,0]
n4 = [0,0,0,0,1]
n5 = [0,0,0,0,0]
matriz = np.array([n1,n2,n3,n4,n5], dtype=bool)
```

```python
print(1*matriz)
```

```python
m2 = matriz@matriz
print(1*m2)
```

```python
m4 = matriz@matriz@matriz@matriz
print(1*m4)
```

```python
for x,y in np.ndenumerate(m4):
  print(x,y)
  if(m4[x]==1):
    print("el nodo {} posee camino de longitud 4 hacia el nodo {}".format(x[0]+1,x[1]+1))
```
