# Estructuras de Control

### Estructuracin por indentacin
* La indentacin es el espacio que existe entre cada lnea de cdigo y el margen.
* A diferencia de muchos otros lenguages de programacin, la indentacin **en python no es slo por esttica sino que define la estructura del cdigo**.

### Toma de decisiones: if .. else ..

La estructura general es:
```python
if condicion:
    # Codigo si la condicion es verdadera
else :
    # Codigo si la condicion es falsa
```

**los bloques se identifican por la indentacin**

### If evala condiciones lgicas

+ puede ser una expresin: 
    
    ``` python 
    if valor > umbral:
        Acciones
    ```
    
+ pueden ser expresiones anidadas: 
    
    ``` python 
    if (valor > piso) and (valor < techo):
        Acciones

    ```
    
    **Ejemplo**
    

````python
texto = 'qweqwe'

if texto == 'Hola':
  print('Hola!, cmo ests?')
else:
  print('...')
````

### Anidando condiciones (elif)

La estructura general es:

```python
if condicion_A:
    #Codigo si la condicion_A es verdadera
elif condicion_B:
    #Codigo si la condicion_B es verdadera
elif condicion_C:
    #Codigo si no cumple ninguna de las anteriores condiciones

```

````python
texto = 'asdasd'

if texto == 'Hola':
  print('Hola!, cmo ests?')
elif texto == 'Chau':
  print('Chau!, nos vemos')
elif texto == 'Buenas Tardes':
  print('Que tal, buenas tardes!')
else:
    print("No se cumplio nada")
````

### List Comprehesion

Hasta ahora vimos **listas** y como podemos recorrerlas con un bucle **for**. 

Python adems nos permite generar una lista de una forma mucho ms consisa y compacta mediante **list comprehesion**.

Por ejemplo:
* Queremos generar una lista de los primeros 10 nmeros pares



Podemos hacerlo utilizando lo aprendido hasta ahora:


````python
lista_num_pares = []

for i in range(0, 11, 3):
    lista_num_pares.append(i)
  
print(lista_num_pares)
````

O mediante list comprehesion

````python
range(10)
````

````python
lista_num_pares = [i**2 for i in range(0, 11, 2)]

print(lista_num_pares)
````
